```python
import pandas as pd
import numpy as np
import seaborn as sns
from tensorflow import keras
```

```python
df = pd.read_csv('Irisdataset.csv')
df.head()
```

```python
df['Species'].value_counts()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df.head()
```

```python
species_name = le.classes_
print(species_name)
```

```python
x = df.drop(columns=['Id', 'Species'])
y = df['Species']
x.head(3)
```

```python
print(y[:5])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, shuffle=True, random_state=69)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
print(x_train[:1])
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print(x_train[:1])
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes=3)
print(y_train[:5])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(x_train.shape[-1], )))
model.add(Dense(units=32,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=3, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, epochs=100, verbose=2)
```

```python
prediction = model.predict(x_test)
```

```python
print(prediction[:5])
```

```python
prediction = np.argmax(prediction, axis=-1)
print(prediction[:5])
```

```python
print(y_test[:5])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, prediction))
```

```python
cm = confusion_matrix(y_test, prediction)
print(cm)
```

```python
ax= sns.heatmap(cm, annot=True, fmt='d' , cmap='Reds', xticklabels=species_name, yticklabels=species_name)
ax.set_title('confusion matrix for irsis dataset prediction')
ax.set_xlabel('prediction', fontsize=14)
ax.set_ylabel('actual, fontsize=14')
```

```python
x_new = np.array([[3, 2, 2, 0.2], [ 4.9, 2.2, 3.8, 1.1 ]])
prediction = model.predict(x_new)
print("Prediction of species: {}".format(species_name))
```
